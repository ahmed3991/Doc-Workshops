#### <span style="color:green"> **3.1-	Averages, sums, counts by category**</span>

```python
# import libraries
import pandas as pd

df_csv = pd.read_csv ('cars.csv')
df = df_csv
df.head() 	# first 5 lines
```

##### *mean()* – **Average of numeric columns**

```python
df.groupby('brand').mean(numeric_only=True)
# This tells Pandas to group by the 'brand' column and t
# hen calculate the mean for all other numeric columns.
```

##### *sum()* – **Total of numeric columns**

```python
df.groupby ('fuel_type')['max_power'].sum()
```

##### *size()* – **Number of rows per group (regardless of column values)**

```python
df.groupby('fuel_type').size()
```

##### *count()* – **Counts non-null values in each column within each group**

```python
df.groupby('brand').count()
```

##### ✏️ **GroupBy Aggregation Methods**
| Method    | What it Does                                        | Example Use Case                            |
| --------- | --------------------------------------------------- | ------------------------------------------- |
| `mean()`  | Calculates the **average** of numeric values        | Average price per brand                     |
| `sum()`   | Calculates the **total sum** of numeric values      | Total power by fuel type                    |
| `size()`  | Counts the **total number of rows** in each group   | Number of cars per transmission type        |
| `count()` | Counts **non-null values** in each column per group | Number of cars with known mileage per brand |


#### <span style="color:green"> **3.2-	Multiple aggregations with agg()**</span>

```python
print(df.columns.tolist())
```

```python
df.groupby('brand').agg({
    'price': ['mean', 'max', 'std'],
    'engine': 'median'   # no spaces in the column name
})
```

#### <span style="color:green"> **3.3-	Aggregation across multiple columns simultaneously**</span>

```python
# This will return the average price per combination of brand and fuel type (like Petrol, Diesel, etc.).
gb = df.groupby(['brand', 'fuel_type'])
gb['price'].mean()
```

#### <span style="color:green"> **3.4-	Pivot Tables**</span>

```python
pd.pivot_table(
    df,
    values='price',
    index='brand',
    columns='fuel_type',
    aggfunc='mean'
)
```

```python
# To fill missing values (e.g., if a brand doesn't offer a fuel type):

pd.pivot_table(
    df,
    values='price',
    index='brand',
    columns='fuel_type',
    aggfunc='mean',
    fill_value=0
)
```

```python
# To round the results:

pivot = pd.pivot_table(df,
                       values='price',
                       index='brand',
                       columns='fuel_type',
                       aggfunc='mean')
pivot.round(0)
```

#### <span style="color:green"> **3.5-	Sorting aggregated groups**</span>

```python
df.groupby('brand')['price'].mean().sort_values(ascending = False)
```

```python
# Re-import libraries after code execution reset
import pandas as pd

# Reload the original cars.csv dataset
df_cars = pd.read_csv('cars.csv')

# Simulated second DataFrame with a different column name for 'brand'
df_extra = pd.DataFrame({
    'car_brand': ['Maruti', 'Hyundai', 'Ford', 'BMW'],  # renamed column
    'avg_service_cost': [5000, 7000, 8000, 12000]
})

# Perform the merge using left_on and right_on
merged_df = pd.merge(df_cars, df_extra, left_on='brand', right_on='car_brand', how='inner')

# Show the first few rows
print(merged_df.head())
```

##### ✏️ **Summary of join types:**
| `how`     | Description                                           |
| --------- | ----------------------------------------------------- |
| `'inner'` | Only matching rows                                    |
| `'left'`  | All rows from the left DataFrame + matches from right |
| `'right'` | All rows from the right DataFrame + matches from left |
| `'outer'` | All rows from both DataFrames                        |


#### <span style="color:green"> **3.7-	Concatenation (stacking multiple arrays)**</span>

```python
#pd.concat([df1,df2],axis = 0 ) 		# line by line
#pd.concat ([df1,df2], axis = 1 ) 		# column by column
```

```python
import pandas as pd

# Load your cars.csv file
df_cars = pd.read_csv('cars.csv')

# Split and concatenate
df1 = df_cars.iloc[:5]
df2 = df_cars.iloc[5:10]

# Row-wise concat (stack rows)
concat_rows = pd.concat([df1, df2], axis=0)
print("Row-wise concatenation:")
print(concat_rows.head(10))

# Column-wise concat (side by side)
concat_cols = pd.concat([df1.reset_index(drop=True), df2.reset_index(drop=True)], axis=1)
print("\nColumn-wise concatenation:")
print(concat_cols.head())
```
